# Creating a new Columns 

```python
profit=[]

#For more relevant data analysis

def result(prof):
    for i in prof:
        if i>=200:
            profit.append('Huge Profit')
        elif i>=50.0 and i<200.0:
            profit.append('Profit')
        elif i>=0.0 and i<50.0:
            profit.append('Margin')
        elif i<0and i>=50.0:
            profit.append('loss')
        else:
            profit.append('Huge Loss')
    return profit
```

```python
result(df.Profit)
```

```python
profit
```

```python
df['Overall']=profit
```

```python
plt.title("Region",fontsize=30)
df['Overall'].value_counts().plot.pie(autopct ='% 1.1f %%', shadow = True)
```

```python
sns.barplot(x="Ship_Mode",y="Sales",hue='Overall',data=df,palette='dark')
```
